```python
#create a series
s = pd.Series([10,20,30,40],index=['a','b','c','d'])
print(s)
#accessing elements with iloc
print(s.iloc[0])    #Accessing the First Element
print(s.iloc[1:3])  #Accessing elements at position 1 and 2(exclusive of 3)
#accessing elements with loc
print(s.loc['a'])     #Accessing the element with index label a 
print(s.loc['b':'c']) ##Accessing the elements with the index labels b and c
```

```python
#Create a DataFrame with multiple Series
data={
    'A':[1,2,3],
    'B':[4,5,6],
    'C':[7,8,9]
}
num_data = pd.DataFrame(data)
print(num_data)
```

```python
#Define a function to sum two series
def sum_series(x,y):
    return x+y
#Apply the function on multiple Series using apply()
result= num_data.apply(lambda row: sum_series(row['A'],row['B']),axis=1) #Adds the  Row A and Row B values
print(result)
```

```python
# List of mobile phone prices
prices = [50000, 40000, 60000, 45000]

# Function to apply a 10% discount
def apply_discount(price):
    return price * 0.9

# Apply the discount to each price using map()
discounted_prices = list(map(apply_discount, prices))

# Print the original and discounted prices
print("Original Prices:", prices)
print("Discounted Prices:", discounted_prices)
```
